```python
import pandas as pd

mse_data = {
    'Dense': mse_results_dense,
    'CNN': mse_results_cnn,
    'Linear': mse_results_linear,
    'LSTM': mse_results_lstm,
    'LSTM-GNN' : mse_results_gnn
}

mse_df = pd.DataFrame(mse_data)

mse_df=mse_df.T
print(mse_df)

```

```python
stock_files = [
    'AAPL', 'MSFT', 'CMCSA', 'COST', 
    'QCOM', 'ADBE', 'SBUX', 'INTU', 
    'AMD', 'INTC'
]
mse_df.columns = stock_files
mse_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

model_names = ['Linear', 'CNN', 'Dense', 'LSTM', 'LSTM-GNN']

sorted_data = sorted(zip(all_mses, model_names), reverse=True)
sorted_mses, sorted_model_names = zip(*sorted_data)

plt.style.use('seaborn-darkgrid')
sns.set_palette("coolwarm", len(sorted_mses))

plt.figure(figsize=(10, 6))
bars = plt.bar(sorted_model_names, sorted_mses, color=sns.color_palette("coolwarm", len(sorted_mses)))

plt.title('Comparison of MSE Across Models (Sorted by MSE)', fontsize=18, fontweight='bold', color='navy')
plt.xlabel('Model', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('MSE', fontsize=14, fontweight='bold', color='darkblue')

plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval * 1.01, round(yval, 5), ha='center', fontsize=12, color='black')

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

mse_df.index = ['Dense', 'CNN', 'Linear', 'LSTM', 'LSTM-GNN']

plt.style.use('ggplot')
sns.set_palette("Blues_d")

for stock in mse_df.columns:
    sorted_mse = mse_df[stock].sort_values(ascending=False)
    plt.figure(figsize=(10, 6))
    bars = plt.bar(sorted_mse.index, sorted_mse.values, color=sns.color_palette("coolwarm", len(sorted_mse)))
    plt.title(f'Model Performance for {stock}', fontsize=16, fontweight='bold', color='navy')
    plt.xlabel('Model', fontsize=14, fontweight='bold', color='darkblue')
    plt.ylabel('MSE', fontsize=14, fontweight='bold', color='darkblue')
    
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.xticks(fontsize=12, fontweight='bold', rotation=45)
    plt.yticks(fontsize=12)
    
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval * 1.01, round(yval, 5), ha='center', fontsize=12, color='black')
    
    plt.subplots_adjust(top=0.85)
    
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
from scipy.stats import ttest_ind
import itertools

model_names = mse_df.index.tolist()

results = []

for model1, model2 in itertools.combinations(model_names, 2):
    group1 = mse_df.loc[model1]
    group2 = mse_df.loc[model2]
    
    t_stat, p_value = ttest_ind(group1, group2, equal_var=False)
    
    results.append({
        'Model 1': model1,
        'Model 2': model2,
        't-statistic': t_stat,
        'p-value': p_value
    })

results_df = pd.DataFrame(results)

print(results_df)
```
